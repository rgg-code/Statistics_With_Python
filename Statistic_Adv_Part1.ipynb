3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(p, x):
  """
  Calculates the probability mass function (PMF) of a Bernoulli distribution.

  Args:
    p: Probability of success.
    x: The value of the random variable (0 for failure, 1 for success).

  Returns:
    The probability of the given x.
  """
  if x == 0:
    return 1 - p
  elif x == 1:
    return p
  else:
    return 0
```

19. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
  """
  Performs a one-sample z-test.

  Args:
    sample_data: A NumPy array containing the sample data.
    population_mean: The hypothesized population mean.
    population_std: The population standard deviation.
    alpha: Significance level (default: 0.05).

  Returns:
    z_statistic: The calculated z-statistic.
    p_value: The p-value of the test.
    reject_null: True if the null hypothesis is rejected, False otherwise.
  """

  sample_mean = np.mean(sample_data)
  sample_size = len(sample_data)
  z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

  # Calculate p-value
  p_value = 2 * (1 - norm.cdf(abs(z_statistic)))  # Two-tailed test

  # Determine whether to reject the null hypothesis
  reject_null = p_value < alpha

  return z_statistic, p_value, reject_null

# Example usage:
sample_data = np.array([102, 98, 105, 97, 101, 103, 99, 100, 104, 98])
population_mean = 100
population_std = 5

z_stat, p_value, reject_null = z_test(sample_data, population_mean, population_std)

print(f"Z-statistic: {z_stat:.2f}")
print(f"p-value: {p_value:.4f}")

if reject_null:
  print("Reject the null hypothesis.")
else:
  print("Fail to reject the null hypothesis.")
```

20. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats

# Sample data
data = np.array([10, 12, 15, 18, 20, 17, 16, 19, 21, 14])

# Calculate the confidence interval (95% confidence level)
confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2)  # Z-score for 95% confidence
standard_error = np.std(data) / np.sqrt(len(data))
margin_of_error = z_critical * standard_error
lower_bound = np.mean(data) - margin_of_error
upper_bound = np.mean(data) + margin_of_error

print(f"95% Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]")

# Interpretation:
# We are 95% confident that the true population mean lies within this interval.
```
